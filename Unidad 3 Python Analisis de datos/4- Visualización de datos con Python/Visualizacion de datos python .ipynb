```python
import pandas as pd
import numpy as np
```

**Paso 1. Carga limpiza y transformacion de tablas**

```python
df_canada = pd.read_excel('Canada.xlsx', sheet_name = 'Canada by Citizenship', skiprows = range(20), skipfooter = 2)
df_canada
```

```python
print (df_canada.shape)
```

```python
df_canada.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace =True)
df_canada.head()
```

```python
df_canada.rename(columns = {'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace = True)
df_canada.head()
```

```python
#Verificar que todos los tipos de comlumnas se sean String 
all(isinstance(column, str) for column in df_canada.columns)
```

```python
#Cambiar todas las columnas a tipo String
df_canada.columns = list(map(str, df_canada.columns))
```

```python
#Verificar que todos los tipos de comlumnas se sean String 
all(isinstance(column, str) for column in df_canada.columns)
```

```python
df_canada.set_index('Country', inplace=True)
df_canada.head()
```

```python
#Agregamos las columns total sumando la migracion 
#de todos los años del pais
df_canada['Total'] = df_canada.sum(axis = 1)
df_canada.head()
```

```python
years = list (map(str, range (1980, 2014)))
years
```

**Parte 2 Visualizacion de datos con matplotlib**

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot')
```

****Graficos de area****

```python
df_canada.sort_values(['Total'], ascending = False, axis = 0, inplace = True)
df_canada.head(5)
```

```python
df_top5 = df_canada.head()
df_top5
```

```python
#Transponer renglones y columnas**
df_top5 = df_top5[years].transpose()
df_top5
```

```python
df_top5.innex = df_top5.index.map(int)
df_top5
```

```python
df_top5.plot (kind = 'area', stacked = False, figsize = (20,10))
plt.title('Tendencias de la inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
#Transparencia
df_top5.plot (kind = 'area', alpha = 0.80 ,stacked = False, figsize = (20,10))
plt.title('Tendencias de la inmigracion de los primeros 5 paises a Canada')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('Años')
plt.show()
```

```python
ax = df_top5.plot(kind = 'area', alpha = 0.35, figsize = (20,10))
ax.set_title('Tendencias de inmigracion de los 5 principales paises a canada')
ax.set_ylabel('Numero de inmigrantes')
ax.set_xlabel('Años')
```

**Grafico de barras**

```python
df_china = df_canada.loc['China', years]
df_china.head()
```

```python
#en los diagramas de barras el eje x se usa para etiquetar y la 
#longitud de la barra corresponde al valor de la variable
df_china.plot(kind = 'bar', figsize = (10,6))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrante chinos a canada desde 1980 al 2013')
plt.show()
```

```python
df_chile = df_canada.loc['Chile', years]
df_chile.head()
```

```python
#en los diagramas de barras el eje x se usa para etiquetar y la 
#longitud de la barra corresponde al valor de la variable
df_chile.plot(kind = 'barh', figsize = (10,6))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrante chilenos a canada desde 1980 al 2013')
for index, value in enumerate (df_chile):
    label = format (int(value), ',')
    plt.annotate(label, xy=(value, index), color='black')
plt.show()
```

```python
df_mexico = df_canada.loc['Mexico', years]
df_mexico.head()
```

```python
#en los diagramas de barras el eje x se usa para etiquetar y la 
#longitud de la barra corresponde al valor de la variable
df_mexico.plot(kind = 'barh', figsize = (10,6))
plt.xlabel('Años')
plt.ylabel('Numero de inmigrantes')
plt.title('Inmigrante mexicanos a canada desde 1980 al 2013')
plt.show()
```

**Graficos circulares**

```python
df_continentes = df_canada.groupby('Continent', axis = 0).sum()
df_continentes
```

```python
explode_list =[0.1, 0,0,0, 0.1,0.3]
df_continentes['Total'].plot(kind='pie',figsize= (15,6), autopct='%1.1f%%', startangle=0, shadow= True, labels=None,pctdistance=1.15, explode = explode_list)

plt.title('Inmigracion a Canada por continente[1980-2013]')
plt.axis('equal')
plt.legend(labels= df_continentes.index, loc='lower left')
plt.show()
```
